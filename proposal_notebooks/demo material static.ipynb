```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import networkx as nx
```

```python
metoo = nx.read_gexf('metoo.gexf')
```

```python
metoo.number_of_edges()
```

```python
labels = nx.get_node_attributes(metoo,'label')
```

```python
edges,weights = zip(*nx.get_edge_attributes(metoo,'weight').items())
```

```python
plt.figure(figsize=(14,16))
nx.draw_spring(metoo,labels=labels, node_color= 'lightpink', font_size = 12,node_size = 100, 
               edge_color= weights, edge_cmap=plt.cm.Blues)
plt.savefig("metoo.png",bbox_inches='tight')
```
