# Updates : 


Findings : 

* Node and Edge both has an alpha parameter that controls the node and edge opacity. Using that it's much easier to get a better graph because we can set the edge opacity lower.

* It's possible to draw labels for only specific nodes. Even without a label adjusting algorithm it's probably possible to have a good visualization if we draw labels for the x largest nodes.



Functions added : 

* ```set_node_color(G,color_by,colormap)``` : Sets the node colors according to the colormap (map from attribute values to colors)

* ```set_node_size(G,size_field,max_node_size)``` : Sets the node sizes according to given node attribute

* ```filter_graph(G,filter_by=None,top=None)```: Filters the graph by given attribute and returns the subgraph from the top x nodes

* ```set_node_label(G,label)``` : Returns the dict containing node labels for the graph from the node attribute that gives the labels

* ```edgecolor_by_source(G,node_colors)``` : returns the list of edge colors colored by the source nodes

* ```get_subgraph_pos(G,pos) ```: Returns the positions for the subgraphs after filtering by some attribute.

````python
from fa2l import force_atlas2_layout
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
G = nx.read_gexf("seth_rich_overall_web.gexf")
G = max(nx.weakly_connected_component_subgraphs(G), key=len).to_undirected()
````

````python
pos = force_atlas2_layout(
        G,
        iterations=50,
        pos_list=None,
        node_masses=None,
        outbound_attraction_distribution=False,
        lin_log_mode=False,
        prevent_overlapping=True,
        edge_weight_influence=1.0,

        jitter_tolerance=1.0,
        barnes_hut_optimize=True,
        barnes_hut_theta=0.5,

        scaling_ratio=20,
        strong_gravity_mode=False,
        multithread=False,
        gravity=1.0)
````

````python
#colormap = {"right":'#f44295','center':'#e5f0f9', 'center_left':'#e5f0f9', 'center_right':'#e5f0f9', 'left':'#e5f0f9', 
#            'null':'#e5f0f9'}
colormap = {"right":'r','center':'g', 'center_left':'g', 'center_right':'r', 'left':'b', 
           'null':'gray'}
color_by = "partisan_retweet"
size_field = 'inlink_count'
````

# Helper Functions 

````python
def set_node_color(G,color_by,colormap=None):
    node_colors = [colormap[n[1][color_by]] for n in G.nodes(data=True)]
    return node_colors
````

````python
def set_node_size(G,size_field,max_node_size):
    max_node_val = max([n[1][size_field] for n in G.nodes(data=True)])
    node_sizes = [max_node_size * (n[1][size_field] / max_node_val) for n in G.nodes(data=True)]
    return node_sizes
````

````python
def filter_graph(G,filter_by=None,top=None):
    filter = {n[0]:n[1][filter_by] for n in G.nodes(data=True)}
    filter_nodes = sorted(filter.items(), key=lambda x: x[1],reverse=True)[0:top]
    return G.subgraph([n[0] for n in filter_nodes])
````

````python
def set_node_label(G,label):
    node_labels = dict((n[0], n[1][label]) for n in G.nodes(data=True))
    return node_labels
````

````python
def edgecolor_by_source(G,node_colors):
    edge_colormap = []
    node_colormap = dict(zip(G.nodes(),node_colors))
    for edge in G.edges():
        edge_colormap.append(node_colormap[edge[0]])
    return edge_colormap

    
````

````python
def get_subgraph_pos(G,pos):
    return {k:v for k,v in pos.items() if k in G.nodes()}
````

# Demo 

````python
top_100_subgraph = filter_graph(G,filter_by="inlink_count",top=100).to_undirected()
````

````python
node_colors = set_node_color(top_100_subgraph,color_by="partisan_retweet",colormap=colormap)
````

````python
node_sizes = set_node_size(top_100_subgraph,size_field= "inlink_count",max_node_size=3000)
````

````python
node_labels = set_node_label(top_100_subgraph,label = "label")
````

````python
subgraph_pos = get_subgraph_pos(top_100_subgraph,pos)
````

````python
edge_colors = edgecolor_by_source(top_100_subgraph,node_colors)
````

````python
top_20_nodes = sorted(zip(top_100_subgraph.nodes(),node_sizes), key= lambda x:x[1], reverse = True)[0:20]
````

````python
top_20_labels = {n[0]:node_labels[n[0]] for n in top_20_nodes}
````
