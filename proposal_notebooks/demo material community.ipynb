```python
import networkx as nx
import community
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
metoo = nx.read_gexf('metoo.gexf')
```

```python
undirected = metoo.to_undirected()
```

```python
partition = community.best_partition(undirected)
```

```python
values = [partition.get(node) for node in metoo.nodes()]

```

```python
labels = nx.get_node_attributes(metoo,'label')
```

```python
edges,weights = zip(*nx.get_edge_attributes(metoo,'weight').items())
```

```python
plt.figure(figsize=(14,16))
nx.draw_spring(metoo,labels=labels, node_color= values, font_size = 12,node_size = 100, 
               edge_color= weights, edge_cmap=plt.cm.cubehelix)
plt.savefig("metoo2.png",bbox_inches='tight')
```
